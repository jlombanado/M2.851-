#### Importamos las librerías

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

#### Se realiza el web scrapping

```python
records = []  

for page in range(0,21,5):
    url = "https://www.tripadvisor.es/Attraction_Review-g187489-d2316449-Reviews-or10-Ayuntamiento_de_Toledo-Toledo_Province_of_Toledo_Castile_La_Mancha.html".format(page)
    html = urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    results = soup.find_all('div', attrs={'class':'prw_rup prw_reviews_basic_review_hsx'})
    for result in results:
        try:
            titulo = result.find('span', attrs={'class':'noQuotes'}).getText()
            opinion = result.find('p', attrs={'class':'partial_entry'}).getText()
            fecha = result.find('span', attrs={'class':'ratingDate relativeDate'}).get("title")
            rate = result.find('div', attrs={'class':'rating reviewItemInline'}).contents[0].get("class")[1]
            records.append((titulo,opinion, rate[7:8], fecha))
        except: AttributeError

 
```

```python
records.
```

#### Se genera el dataframe y el fichero csv resultante

```python
import pandas as pd  
df2 = pd.DataFrame(records, columns=['titulo', 'opinion' , 'rate', 'fecha'])  
df2.to_csv('TripAdvisor.csv', index=False, encoding='utf-8')
```
